```python
import os
from trainer import Trainer
from algorithms.TRPO.agent import TRPOAgent
```

```python
model_name = 'TRPO'
dataset = 'DOW30'
reward = False
print_every = 1
#timestep = 10_000
#trading_cost = 0.0

timesteps = [100_000]
trading_costs = [0.001, 0.01]

for timestep in timesteps:
    for trading_cost in trading_costs:
        if reward:
            strng = 'LogRets'
        else:
            strng = 'ShRt'

        path = f'Reward-{strng}_timestep-{timestep}_trading_cost-{trading_cost}'
        isExist = os.path.exists(path)
        if not isExist:
            os.makedirs(path) 

        name = f'{path}/{model_name}_{path}'

        trainer = Trainer('../data/DOW30.csv', 
                         experiment_name = name, 
                         timesteps = timestep, 
                         print_every = print_every,
                         episode_length = None,
                         returns = reward,
                         trading_cost_ratio = trading_cost ,
                         lookback_period = 64,
                         initial_investment = 1_000_000,
                         random_start_range = 20,
                         dsr_constant = 1e-4,
                         add_softmax = True)

        env = trainer.get_train_env()
        model = TRPOAgent(env, device = 'cpu')
        returns = trainer.run(model)
```
